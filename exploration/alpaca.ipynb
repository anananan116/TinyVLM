```python
import pandas as pd
import numpy as np
import json
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
df = pd.read_json('alpaca_data.json')
df
```

```python
tokenizer = AutoTokenizer.from_pretrained("anananan116/TinyVLM")
```

```python
SYSTEM_PROMPT = "You are a powerful visual assistant."
```

```python
conversationNew = []

for i in range(df.shape[0]):
    conversation = [
        {"role": "system", "content": SYSTEM_PROMPT},
        {"role": "user", "content": df['instruction'][i]},
        {"role": "assistant", "content": df['output'][i]}
    ]
    conversationNew.append(conversation)
```

```python
df['conversation'] = conversationNew
df
```

```python
df.to_csv('alpaca_data_tokenized.csv', index=False)
```
